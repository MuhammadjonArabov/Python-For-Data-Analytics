```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40, 50]
y = [20, 30, 70, 10, 50]

plt.plot(x, y)
plt.title("Line Chart")

plt.title("Line Chart")

plt.ylabel('Y-Axis')

plt.xlabel('X-Axis')
plt.show()
```
