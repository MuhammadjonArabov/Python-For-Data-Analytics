```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ma'lumotlar to'plami
data = {'oquv_soatlari': [5, 10, 3, 8, 12],
        'imtihon_bali': [60, 85, 50, 75, 90]}
df = pd.DataFrame(data)

# Korrelatsiya matritsasi
corr = df.corr()

# Issiqlik xaritasi
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Korrelatsiya xaritasi')
plt.show()

print("Korrelatsiya koeffitsienti:\n", corr)
```
